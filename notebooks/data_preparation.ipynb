```python
import pandas as pd
```

```python
data = pd.read_csv('../data/s1.csv')
```

```python
data.head()
```

```python
# convert GOL, BBH, XCB, NPH, BPL and ZYB to floats

data['GOL'] = data['GOL'].str.replace(',', '.').str.replace('*', '').astype(float)
data['BBH'] = data['BBH'].str.replace(',', '.').str.replace('*', '').astype(float)
data['XCB'] = data['XCB'].str.replace(',', '.').str.replace('*', '').astype(float)
data['NPH'] = data['NPH'].str.replace(',', '.').str.replace('*', '').astype(float)
data['BPL'] = data['BPL'].str.replace(',', '.').str.replace('*', '').astype(float)
data['ZYB'] = data['ZYB'].str.replace(',', '.').str.replace('*', '').astype(float)

data.head()
```

```python
data2 = pd.read_csv('../data/s2.csv')
data2.head()
```

```python
# remove [s<number>] from 'BM (kg) / Reference' create new column BM(kg) and remove [s<number>] from 'ECV (cc) / Reference' create new column ECV(cc)
# use regex to remove the [s<number>] from the column, s followed by any number of digits 

data2['BM (kg)'] = data2['BM (kg) / Reference'].str.replace(r'\[s\d+\]', '', regex=True)
data2['ECV (cc)'] = data2['ECV (cc) / Reference'].str.replace(r'\[s\d+\]', '', regex=True)

# convert BM(kg) and ECV(cc) to floats
data2['BM (kg)'] = data2['BM (kg)'].str.replace(',', '.').str.replace('*', '').astype(float)
data2['ECV (cc)'] = data2['ECV (cc)'].str.replace(',', '.').str.replace('*', '').astype(float)

data2['ID'] = data2['Specimen']

# drop the columns that are not needed
data2 = data2.drop(['Specimen', 'BM (kg) / Reference', 'ECV (cc) / Reference'], axis=1)

data2.head()
```

```python
from statistics import mode

data4['mean'] = pd.to_numeric(data4['mean'], errors='coerce')
# Group by 'ID'
grouped = data4.groupby('ID')

# Calculate the mean of the 'mean' column and the most frequent value in the 'Taxon used' column
result = grouped.agg({'mean': 'mean', 'Taxon used': lambda x: mode(x)})

# Rename the columns for clarity
data4 = result.rename(columns={'mean': 'Mean CC', 'Taxon used': 'Most Frequent Taxon Used'}).reset_index()

# Display the result
data4
```
